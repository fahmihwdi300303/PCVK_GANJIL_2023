```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Load gambar
image = plt.imread('/content/drive/MyDrive/kucing aneh.jpg')
```

```python
# 1. Perhatikan apakah ukuran piksel gambar juga berubah dengan figsize
plt.figure(figsize=(10, 10))
plt.subplot(3, 2, 1)
plt.title("Gambar Asli")
plt.imshow(image)
```

```python
# 3. Tampilkan gambar baris 10-100, kolom 10-100
cropped_image = image[10:101, 10:101, :]
plt.subplot(3, 2, 3)
plt.title("Cropped Gambar (Baris 10-100, Kolom 10-100)")
plt.imshow(cropped_image)
```

```python
# 2. Tampilkan gambar hanya di saluran Merah-Hijau dan Hijau-Biru
red_green_channel = image[:, :, :2]
plt.subplot(3, 2, 2)
plt.title("Saluran Merah-Hijau dan Hijau-Biru")
plt.imshow(red_green_channel)
```

```python
# 4. Tampilkan gambar baris 5-30, semua kolom, saluran Hijau saja
green_channel = image[5:31, :, 1]
plt.subplot(3, 2, 4)
plt.title("Saluran Hijau (Baris 5-30)")
plt.imshow(green_channel, cmap='gray')
```

```python
# 5. Buat 5 kotak dengan ukuran dan warna berbeda dalam satu gambar
num_boxes = 5
colors = ['red', 'green', 'blue', 'purple', 'orange']
random_positions = np.random.randint(0, 255, (num_boxes, 2))
random_sizes = np.random.randint(10, 50, (num_boxes, 2))

plt.subplot(3, 2, 5)
plt.title("Gambar dengan Kotak")
plt.imshow(image)

for i in range(num_boxes):
    x, y = random_positions[i]
    width, height = random_sizes[i]
    rect = patches.Rectangle((x, y), width, height, linewidth=2, edgecolor=colors[i], facecolor='none')
    plt.gca().add_patch(rect)

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
```

```python
# 6. Tunjukkan gambarnya terbalik!
image = cv2.imread('/content/drive/MyDrive/kucing aneh.jpg')
image_inverted = cv2.flip(image, 1)  # Membalik gambar (horizontal)
cv2.imshow('Gambar Terbalik', image_inverted)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 7. Buat persegi panjang dan lingkaran pada bagian muka gambar
height, width, _ = image.shape

# Buat persegi panjang
square_color = (0, 0, 255)  # Warna merah (BGR)
cv2.rectangle(image, (100, 100), (300, 300), square_color, 2)

# Buat lingkaran
circle_center = (width // 2, height // 2)
circle_radius = 50
circle_color = (0, 255, 0)  # Warna hijau (BGR)
cv2.circle(image, circle_center, circle_radius, circle_color, 2)

cv2.imshow('Gambar dengan Persegi Panjang dan Lingkaran', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 8. Buat persegi panjang di pojok kiri bawah saluran B dalam ruang warna RGB
b_channel = image[:, :, 0]  # Saluran B (Blue) dari gambar RGB

# Buat persegi panjang di pojok kiri bawah
square_color_b = 255  # Nilai maksimum untuk saluran B (putih)
cv2.rectangle(b_channel, (10, height - 50), (60, height - 10), square_color_b, -1)

cv2.imshow('Saluran B dengan Persegi Panjang', b_channel)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
